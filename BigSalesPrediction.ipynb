Define Y (dependent or label or target variable) and X (independent or features or attribute variable)

```python
y = df['Item_Outlet_Sales']
```

```python
y.shape
```

```python
y
```

```python
x = df[['Item_Weight', 'Item_Fat_Content', 'Item_Visibility', 'Item_Type', 'Item_MRP',
       'Outlet_Identifier', 'Outlet_Establishment_Year', 'Outlet_Size',
       'Outlet_Location_Type', 'Outlet_Type']]
```

Get X variable standardized

```python
from sklearn.preprocessing import StandardScaler
```

```python
SC = StandardScaler()
```

```python
x_std = df[['Item_Weight', 'Item_Visibility','Item_MRP', 'Outlet_Establishment_Year']]
```

```python
x_std = SC.fit_transform(x_std)
```

```python
x_std
```

```python
x[['Item_Weight', 'Item_Visibility','Item_MRP', 'Outlet_Establishment_Year']] = pd.DataFrame(x_std, columns=['Item_Weight', 'Item_Visibility','Item_MRP', 'Outlet_Establishment_Year'])
```

```python
x
```

Get train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_std_train, x_std_test, y_train, y_test = train_test_split(x_std, y, test_size=0.3, random_state=2529)
```

```python
x_std_train.shape, x_std_test.shape, y_train.shape, y_test.shape
```

Get model train

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=2529)
```

```python
rfr.fit(x_std_train, y_train)
```

Get model prediction

```python
y_pred = rfr.predict(x_std_test)
```

```python
y_pred.shape
```

```python
y_pred
```

Get model evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

Get visualization of actual vs predicted result

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
